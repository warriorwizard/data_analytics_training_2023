# Matplotlib 

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,12,12)
y=x**2
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('title')
plt.plot(x,y,'r-')
# plt.show()
```

# creating multiplot on same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'black')
plt.subplot(1,2,2)
plt.plot(x,y,'r--')
```

# Matplotlib object oriented oriented Method 

```python
fig= plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes = fig.add_axes([])
```
